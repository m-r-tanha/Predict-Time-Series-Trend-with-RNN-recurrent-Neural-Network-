```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense, LSTM, Dropout
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
import math

df = pd.read_excel(r"g:\Desctop 2019.06\TEMP\Time seric\Payload_4year_trend.xlsx",sheet_name='Payload')
input_data= df.iloc[:,1:2].values
sc = MinMaxScaler()
input_data = sc.fit_transform(input_data)
X_train = input_data[0:1493]
y_train = input_data[7:1500]
X_train = np.reshape(X_train,(1493,1,1))

regressor = Sequential()
regressor.add(LSTM(units=4,activation='sigmoid',input_shape=(None,1)))
regressor.add(Dropout(0.1))
regressor.add(Dense(units=1))
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
regressor.fit(X_train, y_train, batch_size = 15, epochs = 500)

test_set = pd.read_excel(r"g:\Desctop 2019.06\TEMP\Time seric\Payload_4year_trend.xlsx",sheet_name='Test')
real_test = test_set.iloc[:,1:2].values

inputs = real_test
inputs = sc.transform(inputs)
inputs = np.reshape(inputs, (176, 1, 1))
predicted = regressor.predict(inputs)
predicted = sc.inverse_transform(predicted)
result=predicted
#result=pd.DataFrame(result)
st=176
'''
result1=[]
for i in np.arange(0,15):

    inputs = np.reshape(result, (176, 1, 1))
    predicted = regressor.predict(inputs)
    predicted = sc.inverse_transform(predicted)
    predicted=pd.DataFrame(predicted)
    result1=result1.append(predicted)
'''

plt.plot(real_test, color = 'red', label = 'Real ')
plt.plot(result, color = 'blue', label = 'Predicted ')
plt.title('Roaming activation day')
plt.xlabel('Time')
plt.ylabel('GByte')
plt.legend()
fig = plt.gcf()
fig.set_size_inches(10, 6)
plt.show()

rmse = math.sqrt(mean_squared_error(real_test, predicted))
print ("Error: "+str(rmse))
```
